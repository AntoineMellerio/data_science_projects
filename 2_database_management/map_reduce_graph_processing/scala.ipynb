### Preprocessing functions

```python
def preprocessing_google(graph):
    graph = graph.zipWithIndex().filter(lambda tup: tup[1] > 3).map(lambda tup: tup[0])
    graph = graph.map(lambda x: x.split('\t')).map(lambda x : (int(x[0]), int(x[1])))
    return graph

#Just taking the header off
def preprocessing_google2(graph):
    graph = graph.zipWithIndex().filter(lambda tup: tup[1] > 3).map(lambda tup: tup[0])
    return graph

def preprocessing_twitter(graph):
    graph = graph.map(lambda x : (x.split(" ")[0],x.split(" ")[1])).map(lambda x : (int(x[0]), int(x[1])))
    return graph
```

### web-Google

Full preprocessing

```python
graph_google = sc.textFile("gs://pjt_bucket1/data1/web-Google.txt.gz")
graph_google2 = preprocessing_google(graph_google)
graph_google2.saveAsSequenceFile('gs://pjt_bucket1/data2/web-Google2.seq')
graph_google2.take(5)
```

Just taking the header off

```python
graph_google = sc.textFile("gs://pjt_bucket1/data1/web-Google.txt.gz")
graph_google2 = preprocessing_google2(graph_google)
graph_google2.saveAsTextFile('gs://pjt_bucket1/data1/web-Google2.txt.gz')
graph_google2.take(5)
```
